### 1. Combine and Clean the Data

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_20_21_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis="rows", join="inner")
athletic_sales_20_21_df
```

```python
# Check if any values are null.
temp_df = athletic_sales_20_21_df.loc[athletic_sales_20_21_df.isnull().any(axis=1)]
temp_df
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_20_21_df['invoice_date'] = pd.to_datetime(athletic_sales_20_21_df['invoice_date'])
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
Product_by_region = athletic_sales_20_21_df.groupby(['region', 'state', 'city'])[['units_sold']].sum().rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
Product_by_region.head()
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
table = pd.pivot_table(athletic_sales_20_21_df, values='units_sold',
                                index=['region', 'state', 'city'], aggfunc='sum').rename(columns={'units_sold': 'Total_Products_Sold'})

# Rename the "units_sold" column to "Total_Products_Sold"

# Show the top 5 results.
table.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
Womens_athletic_shoes = athletic_sales_20_21_df.loc[(athletic_sales_20_21_df['product'] == "Women's Athletic Footwear")]

Retailer_Womens_AS_Tot_sales = Womens_athletic_shoes.groupby(['retailer', 'product'])[['total_sales']].sum().rename(columns={'total_sales': 'Total Sales'}).sort_values(by='Total Sales', ascending=False)

Retailer_Womens_AS_Tot_sales
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
Tot_Retailer_sales_by_region_pvt = pd.pivot_table(Womens_athletic_shoes, values='units_sold',
                                index=['retailer', 'region', 'state', 'city'], aggfunc='sum').rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})


# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

# Show the top 5 results.
Tot_Retailer_sales_by_region_pvt.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
Tot_WAS_by_day_pvt = pd.pivot_table(Womens_athletic_shoes, values='total_sales',
                                index=['invoice_date'], aggfunc='sum').rename(columns={'total_sales': 'Total Sales'}).sort_values(by='Total Sales', ascending=False)


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
Tot_WAS_by_day_pvt
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
Womens_athletic_shoes['weekday'] = (Womens_athletic_shoes['invoice_date']).dt.dayofweek

# Sort the resampled pivot table in ascending order on "Total Sales".
Womens_athletic_shoes_by_day = pd.pivot_table(Womens_athletic_shoes, values='total_sales',
                                index=['weekday'], aggfunc='sum').rename(columns={'total_sales': 'Total Sales'}).sort_values(by='Total Sales', ascending=True)
Womens_athletic_shoes_by_day
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
Womens_athletic_shoes['week_number'] = (Womens_athletic_shoes['invoice_date']).dt.isocalendar().week

Womens_athletic_shoes_by_week = pd.pivot_table(Womens_athletic_shoes, values='total_sales',
                                index=['week_number'], aggfunc='sum').rename(columns={'total_sales': 'Total Sales'}).sort_values(by='Total Sales', ascending=True)
Womens_athletic_shoes_by_week

# Sort the resampled pivot table in ascending order on "Total Sales".
```
